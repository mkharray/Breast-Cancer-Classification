```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Collection and Processing


```python
# loading the breast cancer dataset from sklearn
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
#Loading data to a dataframe
df = pd.DataFrame(dataset.data,columns = dataset.feature_names)
df.head()
```

```python
#adding the diagnosis(label) to the dataframe
#label as in the diagnosis of the particular row -> Malignant(0) and Benign(1)
df['label'] = dataset.target
df.tail()
```

```python
#Info about the dataset
df.info()
```

```python
#Cleaning the data(removing null/missing values)
df.isnull().sum()
```

```python
#Statistical Analysis of the data
df.describe()
```

```python
# 1 -> Benign
# 0 -> Malignant
df['label'].value_counts()
```

```python
#Visual Representation of number of M & B
plt.hist(df['label'], color='g')
plt.title('Plot_Diagnosis (M=0 , B=1)')
plt.show()
```

```python
#Displays the correlation between the labels and the features.
plt.figure(figsize=(40,40))
sns.heatmap(df.corr(), annot=True)
```

```python
#grouping data by their label to compare the different parameters
df.groupby('label').mean()
```

```python
#Seperating the features and target
x = df.drop(columns = 'label',axis = 1) #while dropping rows/columns axis = 1 for dropping columns and axis = 0 for dropping rows
y = df['label']

```

```python
x.shape
```

```python
y.tail()
```

# Splitting Data into Training Data and Test Data

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y, test_size = 0.2, random_state = 2) # test_size -> 20% of data is for testing
```

```python
print(x.shape, X_train.shape, X_test.shape)
```

# Model Training -> Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

# Model Evaluation -> Accuracy Score

```python
prediction = model.predict(X_test)
accuracy = accuracy_score(prediction, Y_test)
print("Accuracy of our Model on testing data = ", (accuracy*100),"%")
```

# Saving the model

```python
import pickle
with open('Breast Cancer Classification Model', 'wb') as f:
  pickle.dump(model,f)
```
